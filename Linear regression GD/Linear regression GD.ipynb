```python
import matplotlib.pyplot as plt
import numpy as np
X = np.arange(200)
delta = np.random.uniform(1,30, size=(200,))

y = .4 * X + 4 + delta
plt.scatter(X, y)
plt.show()
```

### Split dataset

```python
def split_data(X, y, train_ratio=0.8):
    n_samples = X.shape[0]
    n_train = int(n_samples * train_ratio)
    
    # Shuffle the data
    # Split the data into train and test sets
    X_train, X_test = X[:n_train], X[n_train:]
    y_train, y_test = y[:n_train], y[n_train:]
    
    return X_train, X_test, y_train, y_test
```

### Helper Functions (predict, predict_instance)

```python
def predict_instance(X, w, b):
    return (w*X)+b
```

```python
def predict(X, w, b):
    predicts = np.zeros(len(y))
    predicts = (w*X)+b
    return predicts
```

### Errors and accuarcy 

#### Cost

```python
def Cost(X,y,w,b):
    error = 0
    m = len(X)
    for i in range(1,m):
        error+= (y[i] - predict(w,b,X[i]))**2
    return error*((1)/(2*m))
```

#### Loss

#### MSE (Mean square error)

```python
def MSE(y, y_hat):
    return np.mean((y - y_hat)**2)
```

```python
def db(X,y,w,b):
    res = 0.0 
    m = len(X)
    for i in range(m) :
        prediction = predict(X[i],w,b)
        actual = y[i]
        res += (prediction - actual)
    res *= (1/m)
    return res
```

### Gradient descent

```python
def gradient_descent(X, y, learning_rate, n_iters):
    n_samples = len(X)
    
    # Initialize weights and bias to small random values
    weight = np.random.randn()
    bias = np.random.randn()
    
    # Perform gradient descent
    for i in range(n_iters):
        # Compute predictions with current weights and bias
        y_pred = X*weight + bias
        
        # Compute the gradient of the MSE loss function
        grad_weight = np.dot(X.T, (y_pred - y)) / n_samples
        grad_bias = (y_pred - y) / n_samples
        
        # Update weights and bias
        weight -= learning_rate * grad_weight
        bias -= learning_rate * grad_bias
        
        # Print loss every 100 iterations
        if i % 100 == 0:
            loss = MSE(y,y_pred)*(0.5)
            print(f"Iteration {i}: MSE loss = {loss}")
        
        # Check for NaN values
        if np.isnan(weight).any() or np.isnan(bias).any():
            print("Weight or bias contains NaN values, terminating training...")
            break
    
    return weight, bias
```

### Testing

```python
X_train, X_test, y_train, y_test = split_data(X,y)
```

```python
w,b = gradient_descent(X_train, y_train, 0.00001, 40000)
```

```python
print(w,b)
predicts = predict(X_test, w,b)
```
