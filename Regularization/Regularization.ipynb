#### MSE (Mean square error)

```python
def MSE(y, y_hat):
    return np.mean((y - y_hat)**2)
```

### Gradient descent

```python
def gradient_descent(X, y, learning_rate, n_iters):
    n_samples = len(X)
    
    # Initialize weights and bias to small random values
    weight = np.random.randn(3)
    bias = np.random.randn()
    
    # Perform gradient descent
    for i in range(n_iters):
        # Compute predictions with current weights and bias
        y_pred = np.dot(X, weight.T) + bias
        
        # Compute the gradient of the MSE loss function
        grad_weight = np.dot(X.T, (y_pred - y)) / n_samples
        grad_bias = np.mean(y_pred - y)
        
        # Update weights and bias
        weight -= learning_rate * grad_weight
        bias -= learning_rate * grad_bias
        
        # Print loss every 100 iterations
        if i % 100 == 0:
            loss = MSE(y,y_pred)*(0.5)
            print(f"Iteration {i}: MSE loss = {loss}")
        
        # Check for NaN values
        if np.isnan(weight).any() or np.isnan(bias).any():
            print("Weight or bias contains NaN values, terminating training...")
            break
    
    return weight, bias
```
